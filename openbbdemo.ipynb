```python
import openbb
from openbb import obb
obb.user.preferences.output_type = "dataframe"
import numpy as np
```

### Introduction to the OpenBB SDK

OpenBB SDK integrates with multiple data sources (Yahoo Finance, Alpha Vantage, FRED,etc..) allowing you to pull equities, options, forex, crypto and macro-economic data in one place. Depending on the asset type, provider and subscriptions, you can fetch both historical and real time data.

```python
#Use the quote() function to get live stock prices 
real_time_quote = obb.equity.price.quote("AAPL")
print(real_time_quote)
```

```python
obb.crypto.search(query='doge')
obb.crypto.price.historical(symbol='DOGEUSD')
```

```python
obb.currency.price.historical('EURUSD')
```

```python
#historical method accepts a ticker symbol and returns OHLC, volume, split_ratio and dividends in a pandas Dataframe
data = obb.equity.price.historical("NVDA",provider="yfinance")
print(data)
```

```python
#additional parameters for the historical method are start_date, for some providers you can specify the interval, end_date and provider
# see https://docs.openbb.co/platform/reference/equity/price/historical for more details
data = obb.equity.price.historical("NVDA",start_date="2025-01-02",provider="yfinance", interval="1d")
print(data)
```

### Using other data providers
For some methods you can use multiple providers e.g. for historical exchanges rates, the available data providers are FMP,Polygon,Tiingo and YFinance. Some of these data
providers require you to specify the key to be used for accessing their API. These api keys can be specified in the ~/.openbb_platform/user_settings.json file. On a Windows operation system -  this would be c:\windows\users\<currentuser>\.openbb_platform\user_settings.json . For more info see https://docs.openbb.co/platform/user_guides/settings_and_environment_variables 

Some methods are only exposed via specific providers e.g. historical_eps is only available through FMP but there might be additional restrictions  if you are using the free tier which restricts access to specific APIs. 

You can override the key used in a specific session by changing it in Python at runtime usingthe following 2 lines of code

obb.user.credentials.fmp_api_key="<yourapikey>"
obb.account.save()


```python
data = obb.currency.price.historical(symbol='EURUSD', provider='fmp')
print(data)
```

```python
#Insider trading requires access to either fmp or intrinio - FMP free plan does not provide access to insider_trading
#https://docs.openbb.co/platform/reference/equity/ownership/insider_trading
obb.equity.ownership.insider_trading("AAPL",provider='fmp')
```

```python
#FMP free plan is limited to US stocks, you will need to upgrade to a paid plan if you want to retrieve data
#for non-US stocks or ETFs. For US based ETFs you can use SEC as a data provider on holdings but all other methods are exposed using FMP
etfholdings = obb.etf.holdings('URTH',provider='sec')
print(etfholdings)
```
