# Retrieving economic data from FRED

Federal Reserve Economic Data (FRED) - https://fred.stlouisfed.org/  is a database maintained by the Federal Reserve Bank of St. Louis. It has more than 800.000 data time series covering categories such as Economic growth & employment, monetary & fiscal policy, demographics, industries, commodity prices at different frequencies (daily, monthly, annual). It is possible to retrieve this data using the pandas datareader library
 
One of the interesting time series you can find here are 3-month Treasurey Bill Secondary Market rate (TB3MS) or 1-year US Treasury bills which are used a proxy for the risk free rate in financial modeling. There is however some missing data on the TB1YR - so I will be using the TB3MS (3 Month) in my next example. You will notice that are all time series are identified by a short abbreviation that you can find by searching on the FRED website


```python
start = dt.datetime(1980, 5, 2)
end = dt.datetime(2022, 5, 2)
series_code = 'TB3MS' #3 month treasure bill secondary market
data = web.DataReader('TB3MS','fred',start,end)
```
